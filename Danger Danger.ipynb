```python
import pandas as pd
peaks = pd.read_csv("peaks.csv", encoding = 'utf-8')
P1 = peaks.head(50)
peaks.rename(index=str, columns={"pkname": "Name"},inplace=True)
peaks.sort_values(['heightm'],ascending=False,inplace=True)
peaks.reset_index(inplace=True)
peaks.head(5)

```

```python
toppeaks = pd.read_csv("dataminer.csv", encoding = 'utf-8')
toppeaks.drop(toppeaks.index[0:5],inplace=True)
toppeaks['Column 1'].head()
toppeaks.head(3)
```

```python
newpeaks = toppeaks["Column 1"].str.split("/", n = 1, expand = True) 
toppeaks["Latlng"]= newpeaks[0]
toppeaks["Location"]= newpeaks[1]
newpeaks.head()
   
```

```python
newpeaks3 = newpeaks[1].str.split("/", n = 1, expand = True)
newpeaks3.head(2)
```

```python
newpeaks4 = newpeaks3[1].str.split("(", n = 1, expand = True)
print(newpeaks4.head(2))
toppeaks['Location'] = newpeaks4[0]
toppeaks.rename(index=str, columns={"Column 2": "Name"},inplace=True)
toppeaks.head(5)
```

```python
peaks.replace({'Everest': 'Mount Everest'},inplace=True)
peaks.replace({'Kabru main': 'Kabru N'},inplace=True)
peaks.replace({'Himalchuli East': 'Himalchuli'},inplace=True)
peaks.replace({'Peak 29': 'Ngadi Chuli'},inplace=True)
peaks.replace({'Jannu': 'Khumbhakarna'},inplace=True)
peaks.replace({'Jongsang': 'Jongsong Peak'},inplace=True)
peaks.replace({'Ganesh I': 'Yangra'},inplace=True)
peaks.replace({'Gimmigela Chuli': 'Gimmigela'},inplace=True)
peaks.replace({'Pasang Lhamu Chuli': 'Nangpai Gosum'},inplace=True)
peaks.replace({'Churen Himal Central': 'Churen Himal'},inplace=True)
```

```python
toppeaks1 = toppeaks["Name"].str.split("/", n = 1, expand = True)
toppeaks2 = toppeaks1[0].str.split("[", n = 1, expand = True)
toppeaks['Name'] = toppeaks2[0]
toppeaks
peaks1 = peaks.merge(toppeaks,right_on='Name',left_on='Name')
peaks1.drop(['trekyear','trekking','unlisted','pkname2'],axis=1,inplace=True)
Finalpeaks = peaks1.sort_values(['heightm'],ascending=False)
```
