```python
import pandas as pd
members = pd.read_csv("members.csv",encoding = "ISO-8859-1") 
peaks = pd.read_csv("peaks.csv",encoding = "ISO-8859-1",low_memory=False) 
expeditions = pd.read_csv("expeditions.csv",encoding = "ISO-8859-1",low_memory=False) 
```

```python
# Get rid of columns you don't need # 
members = members.drop(columns=['leader','deputy','bconly','nottobc','support','disabled','hired','sherpa','tibetan','msuccess','mclaimed','mdisputed','status','age','birthdate','mtraverse','mparapente','mspeed','mhighpt','mperhighpt','msmtdate1','msmtdate2','msmtdate3','msmttime1','msmttime2','msmttime3','mroute1','mroute2',	'mroute3',	'mascent1',	'mascent2',	'mascent3', 'mo2none','mo2climb','mo2descent','mo2sleep','mo2medical','mo2note' ,'deathdate','deathtime','hcn','mchksum'])
members.head()

```

```python
# Get rid of columns you don't need # 
peaks = peaks.drop(columns=['pkname2','location','himal','region','open','unlisted','trekking','trekyear','restrict','phost','pstatus','psmtnote'])

peaks.head()
```

```python
# Get rid of columns you don't need # 
expeditions = expeditions.drop(columns=['host','route1','route2','route3','route4','nation','leaders','sponsor','success1','success2','success3','success4','ascent1','ascent2','ascent3','ascent4','claimed','disputed','countries','approach','smttime','termdate','termreason','termnote','highpoint','traverse','ski','parapente','rope','nohired','o2none','o2climb','o2descent','o2sleep','o2medical','o2taken','o2unkwn','othersmts','campsites','accidents','achievment','stdrte','primrte','primmem','primref','primid','chksum'])

expeditions.head()
```

```python
members_peaks = pd.merge(members, peaks, on='peakid')
members_peaks.head()
```

```python
expeditions_peaks = pd.merge(expeditions, peaks, on='peakid')
```

```python
#Percentage of Female Alpinists and Total Number#

#Assuming > 12 female summits recorded, does not account for dual citizenship# 

members_peaks['female'] = members_peaks['sex'].map({'F': 1, 'M': 0})
members_peaks['male'] = members_peaks['sex'].map({'F': 0, 'M': 1})

    
members_peaks1 = members_peaks.groupby(['citizen']).sum()

members_peaks2 = members_peaks1[(members_peaks1.female > 12)] 

members_peaks2['Percentage of Female Alpinists'] = members_peaks2['female']/(members_peaks2['female']+members_peaks2['male'])

members_peaks3 = members_peaks2.sort_values(by='Percentage of Female Alpinists', ascending=True)

#members_peaks3.drop(['Australia/New Zealand'])#

FAP = members_peaks3[['Percentage of Female Alpinists','female']]

FAP
```

```python
#How many dual citizens do you have?# 
#What countries are they from?#
members.citizen.dropna()
dual = members.citizen.str.count('/')
dual.sum()

members1 = members[(members.citizen.str.contains('/',na=False))]

membersaus = members1[(members1.citizen.str.contains('Australia'))]

membersaus['sex']

Fem = membersaus.sex.str.count('F')
Mal = membersaus.sex.str.count('M')

Total = Mal.sum() + Fem.sum()
Percentage = Fem.sum()/(Mal.sum() + Fem.sum())

'Australia', Percentage, Total




```
